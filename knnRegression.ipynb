```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
from math import sqrt
import random

# calculating euclidian distance between 2 points

def euclidean_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2
    return sqrt(distance)

# locate k nearest neighbors
def find_neighbor(train, test, k):
    distances = list()
    for row in train:
        dist = euclidean_distance(test,row)
        distances.append((row, dist))
    distances.sort(key=lambda tup: tup[1])
    neighbors = list()

    for i in range(k):
        neighbors.append(distances[i][0])
    return neighbors

# this is my knn regression function
def myknnregress(train, test, k):
    neighbors = find_neighbor(train, test, k)
    print('test point is ',test)
#     print('neighbors are')
#     print(neighbors)
    output_values = [row[-1] for row in neighbors]
    prediction = max(set(output_values), key=output_values.count)
    return neighbors,prediction

      
```
